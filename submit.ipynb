```python
import pandas as pd
from collections import Counter
file_path = 'input_game.csv'
df = pd.read_csv(file_path)

def always_trust(actions):    
    counter_cheat=0  
    for action in actions:
        if(action!='TRUST'):
            counter_cheat+=1
    if(counter_cheat <0.1 * len(actions)):
        return True
    return False
def always_cheat(actions):    
    counter_trust=0  
    for action in actions:
        if(action!='CHEAT'):
            counter_trust+=1
    if(counter_trust <0.1 * len(actions)):
        return True
    return False
def copycat(actions,reaction):
    error=0
    for i in range(1,len(actions)):
        if actions[i]!=reaction[i-1]:
            error+=1
    if error<0.1*len(actions):
        return True
    else:
        return False
def rev_copycat(actions,reaction):
    error=0
    for i in range(1,len(actions)):
        if actions[i]==reaction[i-1]:
            error+=1
    if error<0.1*len(actions):
        return True
    else:
        return False
def simpleton(actions,reaction):
    error=0
    for i in range(1,len(actions)):
        if reaction[i-1]=='TRUST':
            if actions[i]!=actions[i-1]:
                error+=1
        else:
            if actions[i]==actions[i-1]:
                error+=1
                
    if error<0.1*len(actions):
        return True
    else:
        return False
def grudger(actions,reaction):
    error=0
    cheated=False
    if actions[0]=='CHEAT':
        error=1
    if reaction[0]=='CHEAT':
        cheated=True
    for i in range(1,len(actions)):
        if cheated==False:
            if reaction[i-1]=='TRUST' and actions[i]!='TRUST':
                error+=1
            if reaction[i]=='CHEAT':
                cheated=True
        elif cheated==True:
            if actions[i]!='CHEAT':
                error+=1
    if error<0.1*len(actions):
        return True
    else:
        return False
def Trust3Cheat1(actions):
    error=0
    i=0
    for action in actions:
        if i!=3 :
            if action!='TRUST':    
                error+=1
                i=-1
            i+=1
        else:
            if action!='CHEAT':
                error+=1
            i=0
    if error<0.1*len(actions):
        return True
    else:
        return False
def copykitten(actions,reaction):
    error=0
    if actions[0]=='CHEAT':
        error+=1
    if actions[1]=='CHEAT':
        error+=1    
    Moreprev = reaction[0]
    prev = reaction[1]
    for i in range(2,len(actions)):
        if prev == 'CHEAT' and Moreprev=='CHEAT':
            if actions[i]!='CHEAT':
                error+=1
        else:
            if actions[i]!='TRUST':
                error+=1
        Moreprev=prev
        prev=reaction[i]
    if error<0.1*len(actions):
        return True
    else:
        return False

def identify_strategy(actions,reaction):
    if always_trust(actions):
        return 'Always TRUST'
    elif always_cheat(actions):
        return 'Always CHEAT'
    elif rev_copycat(actions, reaction):
        return 'Reverse Copycat '
    elif copycat(actions, reaction):
        return 'CopyCat'
    elif simpleton(actions, reaction):
        return 'Simpleton'
    elif grudger(actions, reaction):
        return 'Grudger'
    elif copykitten(actions, reaction):
        return 'COPYkitten'
    elif Trust3Cheat1(actions):
        return 'Trust3 Cheat1'
    else:
        return 'Other'

# Apply the function to each player
player_strategies = {}
for player_id in df['p1_id'].unique():
    player_actions = df[df['p1_id'] == player_id]['p1_action'].tolist()
    opponent_actions = df[df['p1_id'] == player_id]['p2_action'].tolist()
    # print(len(player_actions))
    # print(len(opponent_actions))
    strategy = identify_strategy(player_actions,opponent_actions)
    player_strategies[player_id] = strategy

# Count the number of players following each strategy
strategy_counts = Counter(player_strategies.values())

# Display the results
print('Number of strategies:', len(strategy_counts))
print('Strategy counts:', strategy_counts)
print('Player strategies:', player_strategies)
```

The above code can be used to identify many strategies like - 

1) Always Trust - In this strategy the player always plays 'TRUST' irrespective of the opponent's move.

2) Always Cheat - This strategy is the converse of the above strategy where the player always plays 'TRUST'. Here the player plays 'CHEAT' move only.

3) CopyCat - In this strategy the player imitates the opponent's last move.

4) Reverse Copycat - In this strategy the player plays the opposite move of the opponent's last move.

5) Simpleton - In this strategy the player imitates his own last move if the opponent's last move is 'TRUST' otherwise the player plays the opposite move of his own last move.

6) Grudger - In this strategy the player plays 'TRUST' until the opponent has played 'CHEAT' atleast once. After which only 'CHEAT' is played by player.

7) CopyKitten - Same as CopyCat just that the player cheats after the opponent has played 'CHEAT' twice consecutively.

8) Trust 3 Cheat 1 - In this the player plays 'TRUST' thrice and 'CHEAT' once in loop.
